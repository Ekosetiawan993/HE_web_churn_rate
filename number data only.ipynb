### We will fill the missing value

How to using value_counts() with nan value

### Region category

```python
train_x.region_category.value_counts(dropna=False)
```

```python
test.region_category.value_counts(dropna=False)
```

```python
train_x.region_category.fillna('Town', inplace=True)
test.region_category.fillna('Town', inplace=True)
```

### Prefered offer types

```python
train_x.preferred_offer_types.value_counts(dropna=False)
```

```python
test.preferred_offer_types.value_counts(dropna=False)
```

```python
train_x.preferred_offer_types.fillna('Gift Vouchers/Coupons', inplace=True)
test.preferred_offer_types.fillna('Gift Vouchers/Coupons', inplace=True)
```

### Points in wallet

```python
train_x.points_in_wallet.value_counts(dropna=False)
```

```python
p_wallet = train_x.points_in_wallet.median()
```

```python
train_x.points_in_wallet.fillna(p_wallet, inplace=True)
test.points_in_wallet.fillna(p_wallet, inplace=True)
```
